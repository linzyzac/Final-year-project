## 1. Capture images from videos

```python
def capture_images_from_videos(video_folder, output_folder, capture_interval=30):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for video_file in os.listdir(video_folder):
        if video_file.endswith(('.mp4', '.avi', '.mkv','.mov','.MOV','.m4v','.MP4')): 
            video_path = os.path.join(video_folder, video_file)
            
            cap = cv2.VideoCapture(video_path)
            
            fps = cap.get(cv2.CAP_PROP_FPS)
            total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            
            frame_interval = int(fps * capture_interval)
            
            for i in range(0, total_frames, frame_interval):
                cap.set(cv2.CAP_PROP_POS_FRAMES, i)
                ret, frame = cap.read()
                if ret:
                    image_filename = f"{video_file}_frame_{i}.jpg"
                    image_path = os.path.join(output_folder, image_filename)
                    cv2.imwrite(image_path, frame)
            
            cap.release()

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part1/01'  
    output_folder = './uta_dataset/output_images/output_images1/01'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part1/02'  
    output_folder = './uta_dataset/output_images/output_images1/02'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part1/03'  
    output_folder = './uta_dataset/output_images/output_images1/03'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part1/04'  
    output_folder = './uta_dataset/output_images/output_images1/04'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part1/05'  
    output_folder = './uta_dataset/output_images/output_images1/05'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part1/06'  
    output_folder = './uta_dataset/output_images/output_images1/06'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part2/07'  
    output_folder = './uta_dataset/output_images/output_images1_2/07'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part2/08'  
    output_folder = './uta_dataset/output_images/output_images1_2/08'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part2/09'  
    output_folder = './uta_dataset/output_images/output_images1_2/09'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part2/10'  
    output_folder = './uta_dataset/output_images/output_images1_2/10'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part2/11'  
    output_folder = './uta_dataset/output_images/output_images1_2/11'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold1_part2/12'  
    output_folder = './uta_dataset/output_images/output_images1_2/12'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
    
if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part1/13'  
    output_folder = './uta_dataset/output_images/output_images2/13'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part1/14'  
    output_folder = './uta_dataset/output_images/output_images2/14'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part1/15'  
    output_folder = './uta_dataset/output_images/output_images2/15'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part1/16'  
    output_folder = './uta_dataset/output_images/output_images2/16'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part1/17'  
    output_folder = './uta_dataset/output_images/output_images2/17'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part1/18'  
    output_folder = './uta_dataset/output_images/output_images2/18'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part2/19'  
    output_folder = './uta_dataset/output_images/output_images2_2/19'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part2/20'  
    output_folder = './uta_dataset/output_images/output_images2_2/20'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part2/21'  
    output_folder = './uta_dataset/output_images/output_images2_2/21'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part2/22'  
    output_folder = './uta_dataset/output_images/output_images2_2/22'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part2/23'  
    output_folder = './uta_dataset/output_images/output_images2_2/23'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold2_part2/24'  
    output_folder = './uta_dataset/output_images/output_images2_2/24'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part1/25'  
    output_folder = './uta_dataset/output_images/output_images3/25'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part1/26'  
    output_folder = './uta_dataset/output_images/output_images3/26'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part1/27'  
    output_folder = './uta_dataset/output_images/output_images3/27'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part1/28'  
    output_folder = './uta_dataset/output_images/output_images3/28'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part1/29'  
    output_folder = './uta_dataset/output_images/output_images3/29'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part1/30'  
    output_folder = './uta_dataset/output_images/output_images3/30'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part2/31'  
    output_folder = './uta_dataset/output_images/output_images3_2/31'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part2/32'  
    output_folder = './uta_dataset/output_images/output_images3_2/32'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part2/33'  
    output_folder = './uta_dataset/output_images/output_images3_2/33'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part2/34'  
    output_folder = './uta_dataset/output_images/output_images3_2/34'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part2/35'  
    output_folder = './uta_dataset/output_images/output_images3_2/35'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold3_part2/36'  
    output_folder = './uta_dataset/output_images/output_images3_2/36'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part1/37'  
    output_folder = './uta_dataset/output_images/output_images4/37'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part1/38'  
    output_folder = './uta_dataset/output_images/output_images4/38'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part1/39'  
    output_folder = './uta_dataset/output_images/output_images4/39'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part1/40'  
    output_folder = './uta_dataset/output_images/output_images4/40'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part1/41'  
    output_folder = './uta_dataset/output_images/output_images4/41'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part2/42'  
    output_folder = './uta_dataset/output_images/output_images4_2/42'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part2/43'  
    output_folder = './uta_dataset/output_images/output_images4_2/43'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part2/44'  
    output_folder = './uta_dataset/output_images/output_images4_2/44'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part2/45'  
    output_folder = './uta_dataset/output_images/output_images4_2/45'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part2/46'  
    output_folder = './uta_dataset/output_images/output_images4_2/46'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part2/47'  
    output_folder = './uta_dataset/output_images/output_images4_2/47'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)

if __name__ == "__main__":
    video_folder = './uta_dataset/UTA/Fold4_part2/48'  
    output_folder = './uta_dataset/output_images/output_images4_2/48'     
    capture_interval = 5               

    capture_images_from_videos(video_folder, output_folder, capture_interval)
```

## 2. Traditional Machine Learning Data Preprocessing

## 2.1 Resize images

```python
def detect_and_crop(image_path, output_path, target_size=(512, 512), quality=85):
    # Load the image using OpenCV
    img = cv2.imread(image_path)
    
    # Convert the image to grayscale for face detection
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Load the pre-trained face detector from OpenCV
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    if len(faces) > 0:
        # Get the coordinates of the first detected face
        x, y, w, h = faces[0]

        # Crop the image to focus on the detected face
        face_img = img[y:y+h, x:x+w]

        # Resize the cropped image to the target size
        resized_img = cv2.resize(face_img, target_size)

        # Save the resized image to the output path
        Image.fromarray(resized_img).save(output_path, quality=quality)

if __name__ == "__main__":
    input_folder = "./uta_dataset/output_images/uta_dataset_images/Alert"
    output_folder = "./uta_dataset/final_uta_dataset_images/Alert"
    target_size = (512, 512)
    quality = 85

    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Process each image in the input folder
    for file in os.listdir(input_folder):
        if file.lower().endswith(('.png', '.jpg', '.jpeg','.JPG', 'JPEG', '.PNG')):
            input_path = os.path.join(input_folder, file)
            output_path = os.path.join(output_folder, file)
            detect_and_crop(input_path, output_path, target_size, quality)


if __name__ == "__main__":
    input_folder = "./uta_dataset/output_images/uta_dataset_images/Drowsy"
    output_folder = "./uta_dataset/final_uta_dataset_images/Drowsy"
    target_size = (512, 512)
    quality = 85

    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Process each image in the input folder
    for file in os.listdir(input_folder):
        if file.lower().endswith(('.png', '.jpg', '.jpeg','.JPG', 'JPEG', '.PNG')):
            input_path = os.path.join(input_folder, file)
            output_path = os.path.join(output_folder, file)
            detect_and_crop(input_path, output_path, target_size, quality)
```

## 2.2 Get facial landmarks

```python
bz2_file_path = './shape_predictor_68_face_landmarks.dat.bz2'
decompressed_file_path = './shape_predictor_68_face_landmarks.dat'
faces_folder_path = './uta_dataset/final_uta_dataset_images/Alert'

with bz2.BZ2File(bz2_file_path, 'rb') as f_in, open(decompressed_file_path, 'wb') as f_out:
    shutil.copyfileobj(f_in, f_out)

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(decompressed_file_path)
# win = dlib.image_window()

def shape_to_np(shape, dtype="int"):
	# initialize the list of (x, y)-coordinates
	coords = np.zeros((68, 2), dtype=dtype)
	# loop over the 68 facial landmarks and convert them
	# to a 2-tuple of (x, y)-coordinates
	for i in range(0, 68):
		coords[i] = (shape.part(i).x, shape.part(i).y)
	# return the list of (x, y)-coordinates
	return coords

def rect_to_bb(rect):
	# take a bounding predicted by dlib and convert it
	# to the format (x, y, w, h) as we would normally do
	# with OpenCV
	x = rect.left()
	y = rect.top()
	w = rect.right() - x
	h = rect.bottom() - y
	# return a tuple of (x, y, w, h)
	return (x, y, w, h)

# for f in glob.glob(os.path.join(faces_folder_path, "*.png","*.jpg"))[0:1]:
#     print("Processing file: {}".format(f))
    
# load the input image, resize it, and convert it to grayscale
image = cv2.imread('./uta_dataset/final_uta_dataset_images/Drowsy/10.MOV_frame_0.jpg')
image = imutils.resize(image, width=500)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rects = detector(gray, 1)
# print("Number of faces detected: {}".format(len(dets)))
# loop over the face detections
for (i, rect) in enumerate(rects):
    # determine the facial landmarks for the face region, then
    # convert the facial landmark (x, y)-coordinates to a NumPy
    # array
    shape = predictor(gray, rect)
    shape = shape_to_np(shape)
    # convert dlib's rectangle to a OpenCV-style bounding box
    # [i.e., (x, y, w, h)], then draw the face bounding box
    (x, y, w, h) = rect_to_bb(rect)
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    # show the face number
    cv2.putText(image, "Face #{}".format(i + 1), (x - 10, y - 10),
        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    # loop over the (x, y)-coordinates for the facial landmarks
    # and draw them on the image
    for (x, y) in shape:
        cv2.circle(image, (x, y), 5, (255, 0, 0), -1)
# show the output image with the face detections + facial landmarks
# print(image.shape)
#Show the image with matplotlib
plt.imshow(image)
plt.show()
```

## 2.3 Landmarks to Microsoft Excel

```python
bz2_file_path = 'shape_predictor_68_face_landmarks.dat.bz2'
decompressed_file_path = 'shape_predictor_68_face_landmarks.dat'

with bz2.BZ2File(bz2_file_path, 'rb') as f_in, open(decompressed_file_path, 'wb') as f_out:
    shutil.copyfileobj(f_in, f_out)

def get_landmarks(image_path):
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(decompressed_file_path)

    img = dlib.load_rgb_image(image_path)
    dets = detector(img)
    
    if not dets:
        return None
    
    shape = predictor(img, dets[0])
    landmarks = [(shape.part(i).x, shape.part(i).y) for i in range(shape.num_parts)]

    return landmarks

def save_landmarks_to_excel(input_folder, output_excel):
    data = {'Image': [], 'Landmarks': [], 'label':[]}

    for image_file in os.listdir(input_folder):
        if image_file.endswith(('.jpg', '.png', '.jpeg','.JPG','.JPEG','.PNG')):
            image_path = os.path.join(input_folder, image_file)
            
            landmarks = get_landmarks(image_path)
            
            if landmarks is not None:
                data['Image'].append(image_file)
                data['Landmarks'].append(landmarks)
                data['label'].append(1)

    df = pd.DataFrame(data)
    df.to_excel(output_excel, index=False)


if __name__ == "__main__":
    input_folder = './uta_dataset/final_uta_dataset_images/Drowsy'
    output_excel = './landmarks/landmarks_data_drowsy.xlsx'

    save_landmarks_to_excel(input_folder, output_excel)
    
if __name__ == "__main__":
    input_folder = './uta_dataset/final_uta_dataset_images/Alert'
    output_excel = './landmarks/landmarks_data_alert.xlsx'

    save_landmarks_to_excel(input_folder, output_excel)
```

## 3.4 Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

import graphviz
import os
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print("Precision:",metrics.precision_score(y_test, y_pred))

print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
for i in range(3):
    X_train_array = np.array(X_train)
    tree = rf.estimators_[i]
    dot_data = export_graphviz(tree,
                               feature_names=['feature_' + str(i) for i in range (X_train_array.shape[1])],  
                               filled=True,
                               rounded=True,
                               special_characters=True,
                               max_depth=2,
                               class_names=['Class 0','Class 1'])
    graph = graphviz.Source(dot_data)
    display(graph)
```

```python
param_dist = {'n_estimators': randint(50,500),
              'max_depth': randint(1,20)}

rf = RandomForestClassifier()

rand_search = RandomizedSearchCV(rf, 
                                 param_distributions = param_dist, 
                                 n_iter=5, 
                                 cv=5)

rand_search.fit(X_train, y_train)
```

```python
best_rf = rand_search.best_estimator_

print('Best hyperparameters:',  rand_search.best_params_)
```

```python
y_pred = best_rf.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

ConfusionMatrixDisplay(confusion_matrix=cm).plot();
```
